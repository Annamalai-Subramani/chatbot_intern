```python
import os
import rasa
import nest_asyncio
from rasa.cli.scaffold import create_initial_project
from rasa.jupyter import chat

nest_asyncio.apply()

# Create a new Rasa project
project = 'testproject'
create_initial_project(project)
os.chdir(project)

# Print the contents of the project directory
print(os.listdir('.'))


```

```python
# Define paths
config = 'config.yml'
training_files = 'data/'
domain = 'domain.yml'
output = 'models/'

# Train the model
training_result = rasa.train(domain, config, [training_files], output)
model_path = training_result.model

# Print the path to the trained model
print("Model trained successfully. Model path:", model_path)

# Start a chat session
chat(model_path)
```

```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

# Sample dataset (replace this with your actual dataset)
training_data = [
    ("I want to book a hotel room", {"intent": "book_room", "entities": {"location": "Paris"}}),
    ("Book a room for me", {"intent": "book_room", "entities": {"location": "London"}}),
    ("I need to cancel my reservation", {"intent": "cancel_resr", "entities": {}}),
    # Add more examples with intents and entities
]

# Train the intent recognition model
def train_intent_recognition(data):
    texts = [item[0] for item in data]
    intents = [item[1]["intent"] for item in data]

    # Use CountVectorizer for text representation
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(texts)

    # Train a Support Vector Classifier for intent classification
    classifier = SVC(kernel='linear')
    classifier.fit(X, intents)

    return Pipeline([
        ('vectorizer', vectorizer),
        ('classifier', classifier)
    ])

# Test intent recognition model
intent_recognition_model = train_intent_recognition(training_data)
test_texts = ["I want to book a room", "Cancel my reservation"]
predicted_intents = intent_recognition_model.predict(test_texts)
print("Predicted intents:", predicted_intents)

# Named Entity Recognition (NER) using spaCy
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    entities = {ent.label_: ent.text for ent in doc.ents}
    return entities

# Test entity extraction
text = "Book a room in Paris for two nights"
entities = extract_entities(text)
print("Entities:", entities)

# Evaluation and iterative improvement
# Evaluate the performance of the intent recognition model using classification report
def evaluate_intent_recognition(model, data):
    texts = [item[0] for item in data]
    true_intents = [item[1]["intent"] for item in data]
    predicted_intents = model.predict(texts)
    print("Classification Report:")
    print(classification_report(true_intents, predicted_intents))

# Test evaluation
evaluate_intent_recognition(intent_recognition_model, training_data)

# Documentation and reporting: Document the development process and submit the assignment
# Include code comments, explanations, and results in the documentation
```
